# Silme İşlemi

```python
import pandas as pd
import numpy as np
```

```python
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df = pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3}        
)

df
```

```python
df.dropna()
```

```python
df.dropna(how="all")
```

```python
df.dropna(axis=1)
```

```python
df.dropna(axis=1,how="all")
```

```python
df["sil_beni"]=np.nan
df
```

```python
df.dropna(axis=1,how="all",inplace=True)
df
```
