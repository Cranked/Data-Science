### Veri Standardizasyonu

#### Standartlaştırma

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
V1 = np.array([1,3,6,7,1,9])
V2 = np.array([7,5,8,12,2,3])
V3 = np.array([12,5,6,14,7,5])
df = pd.DataFrame(
        {
         "V1" : V1,
         "V2" : V2,
         "V3" : V3
        }        
)
df=df.astype(float)
df
```

#### Normalizasyon

```python
preprocessing.scale(df)
```

#### Normalizasyon

```python
preprocessing.normalize(df)
```

### Min-Max Dönüşümü

```python
scaler=preprocessing.MinMaxScaler(feature_range=(10,20))
```

```python
scaler.fit_transform(df)
```

#### Binarize Dönüşüm

```python
binarizer=preprocessing.Binarizer(threshold=5).fit(df)
```

```python
binarizer.transform(df)
```

#### 0-1 Dönüşümü

```python
import seaborn as sns
tips=sns.load_dataset("tips")
df=tips.copy()
df_l=df.copy()
df_l.head()
```

```python
df_l["yeni_sex"]=df_l.sex.cat.codes
df_l
```

```python
lbe=preprocessing.LabelEncoder()
```

```python
df_l["daha_yeni_sex"]=lbe.fit_transform(df_l.sex)
df_l.head()
```

#### "1 ve  Diğerleri(0) " Dönüşümü

```python
df_l["yeni_gun"]=np.where(df.day.str.contains("Sat"),1,0)
df_l.head(20)
```

### Çok sınıflı Dönüşüm

```python
df_l["daha_yeni_gun"]=lbe.fit_transform(df_l.day)
df_l
```
